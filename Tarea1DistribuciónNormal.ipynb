#La altura de una población sigue una distribución normal con media μ = 160 cm y desviación estándar σ = 10 cm.

```python
from scipy.stats import norm
import numpy as np

# Parámetros
mu = 160  # media
sigma = 10  # desviación estándar
x_values = np.linspace(mu - 3*sigma, mu + 3*sigma, 20)  # rango de valores

# Generar la tabla de probabilidades
tabla_probabilidades = [(x, norm.pdf(x, mu, sigma)) for x in x_values]

# Imprimir la tabla
print("Tabla de probabilidades:")
print("x\tProbabilidad")
for x, probabilidad in tabla_probabilidades:
    print(f"{x:.2f}\t{probabilidad:.4f}")
```

a) ¿Cuál es la probabilidad de que una persona seleccionada al azar
tenga una altura mayor a 170 cm?

```python
from scipy.stats import norm

# Parámetros
mu = 160  # media
sigm = 10  # desviación estándar
x = 170  # valor específico

# Calcular la probabilidad acumulada
probabilidad_mas_de_x = 1 - norm.cdf(x, mu, sigma)
print(f"La probabilidad de que la variable sea mayor que {x} es {probabilidad_mas_de_x:.4f}")
```

b) ¿Y cuál es la probabilidad de que la altura sea menor o igual a 155
cm?

```python
from scipy.stats import norm

# Parámetros
mu = 160  # media
sigma = 10  # desviación estándar
x = 155  # valor específico

# Calcular la probabilidad acumulada
probabilidad_acumulada = norm.cdf(x, mu, sigma)
print(f"La probabilidad de que la variable sea menor o igual a {x} es {probabilidad_acumulada:.4f}")
```

c) Encuentra la probabilidad de que una persona tenga una altura entre
150 cm y 165 cm.

```python
from scipy.stats import norm

# Parámetros
mu = 160  # media
sigma = 10  # desviación estándar
x1 = 150  # límite inferior
x2 = 165  # límite superior

# Calcular la probabilidad
probabilidad_entre_x1_y_x2 = norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma)
print(f"La probabilidad de que la variable esté entre {x1} y {x2} es {probabilidad_entre_x1_y_x2:.4f}")
```

d) Si el 20 % de las personas más altas son consideradas muy altas, ¿cuál
es la altura a partir de la cual una persona puede ser considerada muy
alta?

```python
from scipy.stats import norm

# Parámetros de la distribución
mu = 160      # media
sigma = 15    # desviación estándar

# Queremos el punto donde queda arriba el 20% de la población
# Es decir el percentil 80 (0.80)
percentil_80 = norm.ppf(0.80, mu, sigma)

print(f"La altura mínima para ser considerada muy alta es: {percentil_80:.2f} cm")
```
