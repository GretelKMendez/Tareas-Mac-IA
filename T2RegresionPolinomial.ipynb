#Consumo Energético.

#Consumo de energia con temperatura en 32°

```python
# Regresión Polinómica en Python

# Sección 1: Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Sección 2: Datos de ejemplo
X = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55]).reshape(-1, 1)
y = np.array([200, 230, 270, 300, 340, 380, 420, 460, 500, 550])

# Sección 3: Crear y entrenar modelos polinómicos de distintos grados
grados = [1, 2, 4]
modelos = []
predicciones = []

for grado in grados:
    modelo = make_pipeline(PolynomialFeatures(degree=grado), LinearRegression())
    modelo.fit(X, y)
    modelos.append(modelo)
    y_pred = modelo.predict(X)
    predicciones.append(y_pred)
    print(f"\nModelo de grado {grado}:")
    print(f"MSE: {mean_squared_error(y, y_pred):.4f}")
    print(f"R^2: {r2_score(y, y_pred):.4f}")

# Sección 4: Visualización de los ajustes
X_plot = np.linspace(10, 55, 100).reshape(-1, 1)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='red', label='Datos reales')

colores = ['blue', 'green', 'orange']

for i, grado in enumerate(grados):
    y_plot = modelos[i].predict(X_plot)
    plt.plot(X_plot, y_plot, color=colores[i], label=f"Grado {grado}")

plt.title("Regresión Polinómica de distintos grados")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Sección 5: Predicción para nuevo valor
nuevo_x = np.array([[32]])
for i, grado in enumerate(grados):
    pred = modelos[i].predict(nuevo_x)
    print(f"Predicción para x=32 con grado {grado}: {pred[0]:.2f}")
```
